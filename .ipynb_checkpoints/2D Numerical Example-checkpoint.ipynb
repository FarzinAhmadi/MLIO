```python
def twoD_example_random_generator (A,b,centers,num_points,figure):
    n = np.sum(num_points)
    X = np.zeros ([2,n])
    count = 0
    std_dev = 0.7
    for i in range(centers.shape[0]):
        X[0][count:count + num_points[i]] = std_dev* np.random.randn(1,num_points[i])+ centers.T[0,i]
        X[1][count:count + num_points[i]] = std_dev* np.random.randn(1,num_points[i])+ centers.T[1,i]
        count = count + num_points[i]
    if figure == True:
        plt.figure(figsize=(7,7))
        s = 35
        plt.scatter(X[0, :], X[1, :], c="black", s=30, edgecolor="black", label="train_data")
        #plt.scatter(X_test[0, :], X_test[1, :], c="red", s=30, edgecolor="red", label="Test_data")
        plt.xlim([0, 11])
        plt.ylim([0, 11])
        plt.xlabel(u'x\u2081')
        plt.ylabel(u'x\u2082')
        plt.plot([1,1], [1, 10], c="black")
        plt.plot([10,10], [1, 5], c="black")
        plt.plot([1,5], [10, 10], c="black")
        plt.plot([1,10], [1, 1], c="black")
        plt.plot([10,5], [5, 10], c="black")
        plt.fill_between([1,5], [10,10],[1,1], facecolor='black', alpha=0.15)
        plt.fill_between([5,10], [10,5],[1,1], facecolor='black', alpha=0.15)
    return X


A = np.array (([1,0]
              ,[-1,0]
              ,[0,1]
              ,[0,-1]
              ,[1,1]))
b = np.array ([10,-1,10,-1,15])
centers = np.array([[2,5]
                   ,[6,2]
                   ,[8.5,5]
                   ,[5.5,9]])
num_points = np.array([25,25,25,25])
figure = True
X= twoD_example_random_generator (A,b,centers,num_points,True)
X_train = X.T
```
